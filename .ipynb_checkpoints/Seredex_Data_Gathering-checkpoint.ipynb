## Function Definitions

```python
def fun_get_soup(html, user_agent_string = 'Electronix Goddess'):
    temp_get = requests.get(html, headers = {'User-agent' : user_agent_string})
    return BeautifulSoup(temp_get.text, 'lxml')
```

```python
def fun_validate_name(html):
    return "Pok%C3%A9mon" in html
```

```python
def fun_test(soup):
    temp_list = [x for x in soup.select("h3 + table")][1]

    temp_df = pd.read_html(str(temp_list))[0]

    return temp_df#[~temp_df[2].isna()]
```
